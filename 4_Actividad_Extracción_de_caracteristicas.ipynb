```python
#Tambien podremos verificar la cantidad de datos nulos que
#faltan en las variables
valores_nulos = Filtrados.isnull().sum()
valores_nulos
```

```python
#Dependiendo del criterio se aplica la limpieza de datos
Filtrados['184_store_devices'] = Filtrados['184_store_devices'].fillna(method = 'bfill')
Filtrados['5_change_store_space_last_year'] = Filtrados['5_change_store_space_last_year'].fillna(method = 'ffill')
Filtrados['6_change_employees_average_salary_last_year'] = Filtrados['6_change_employees_average_salary_last_year'].fillna(method = 'ffill')
Filtrados['49_inventory_records'] = Filtrados['49_inventory_records'].fillna(method = 'ffill')
Filtrados['18_sales_records'] = Filtrados['18_sales_records'].fillna(method = 'ffill')
Filtrados['210_sales_channels'] = Filtrados['210_sales_channels'].fillna(method = 'ffill')
Filtrados['189_payment_methods'] = Filtrados['189_payment_methods'].fillna(method = 'ffill')
Filtrados['214_customer_relationship_tools'] = Filtrados['214_customer_relationship_tools'].fillna(method = 'ffill')
Filtrados['193_sales_planning_tools'] = Filtrados['193_sales_planning_tools'].fillna(method = 'ffill')
Filtrados['60_services'] = Filtrados['60_services'].fillna(method = 'ffill')
Filtrados['60_services'] = Filtrados['60_services'].fillna(method = 'bfill')
Filtrados['42_willingness_share_inventory_data_suppliers'] = Filtrados['42_willingness_share_inventory_data_suppliers'].fillna(method = 'ffill')
Filtrados['42_willingness_share_inventory_data_suppliers'] = Filtrados['42_willingness_share_inventory_data_suppliers'].fillna(method = 'bfill')
Filtrados['45_willingness_reorder_suppliers'] = Filtrados['45_willingness_reorder_suppliers'].fillna(method = 'ffill')
Filtrados['45_willingness_reorder_suppliers'] = Filtrados['45_willingness_reorder_suppliers'].fillna(method = 'bfill')
Filtrados['158_charact_organize_products_on_shelves'] = Filtrados['158_charact_organize_products_on_shelves'].fillna(method = 'ffill')
Filtrados['161_actions_stockouts'] = Filtrados['161_actions_stockouts'].fillna(method = 'ffill')
Filtrados['249_technology_training'] = Filtrados['249_technology_training'].fillna(method = 'ffill')
Filtrados['249_technology_training'] = Filtrados['249_technology_training'].fillna(method = 'bfill')
Filtrados['250_technology_investment_operations'] = Filtrados['250_technology_investment_operations'].fillna(method = 'ffill')
Filtrados['250_technology_investment_operations'] = Filtrados['250_technology_investment_operations'].fillna(method = 'bfill')
```
